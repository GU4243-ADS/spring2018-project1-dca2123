### Sentiment Analysis

Now, we will compare how positive, negative, or neutral each author is. Although these author wrote at least over 100 years ago, positiveity and negativity 
are fundamental, basic concepts in human communication. The words communicating those sentiments are more likely to remain the same or similar over time.

With that said, let's get started.

```python
# inspired by http://www.nltk.org/howto/sentiment.html
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def avg_polar_score(sentences,sia):
    '''Takes some sentences and a SentimentIntensityAnalyzer object and returns average of their polarity scores'''
    n = sentences.size
    scores = {"neg" : 0,"neu" : 0,"pos" : 0,"compound" : 0}
    
    for sentence in sentences:
        #get the sia polarity score for a single sentence.
        ss = sia.polarity_scores(sentence)
        
        #now add those scores to all the other scores that have been calculated
        for k in sorted(ss):
            #add each score individually within the dict
            scores[k] += ss[k]
    
    #average those totals
    for j in sorted(scores):
        scores[j] = round((scores[j]/n),3)
    
    return(scores)
```

```python
some_sia = []

some_sia.append(avg_polar_score(eap.text.as_matrix(),SentimentIntensityAnalyzer()))
some_sia.append(avg_polar_score(hpl.text.as_matrix(),SentimentIntensityAnalyzer()))
some_sia.append(avg_polar_score(mws.text.as_matrix(),SentimentIntensityAnalyzer()))
```

Simple enough code, and the output below provides a couple clear differences about the authors:

EAP and HPL stray from neutral about the same amount on average.
MWS, comparitively, swings through positive and negative sentiments far more often.

EAP and MWS tend to be more positive than negative on average; HPL favor negativity

```python
author_names = ['Edgar Allen Poe','H.P. Lovecraft','Mary Wolstencraft Shelly']
i = 0
for dic in some_sia:
    print(author_names[i])
    i += 1
    for k in sorted(dic):
        print('{0}: {1}, '.format(k, dic[k]), end='')
    print('\n')
```

## Topic Modeling

I have chosen to highlight the differences in the results of the Latent Dirilct Allocation (LDA) algorithm, and Non-negative Matrix Factorization (NMF)
algorithm.

I will focus mainly on the difference between LDA and NMF with MWS, because 1. this phase of the analysis is almost purely subjective and my
analysis and bias would be redundant if indulged further, and 2. MWS had the most convincing topics.

```python
#topic modeling code inspired by https://medium.com/mlreview/topic-modeling-with-scikit-learn-e80d33668730
#the idea to compare lda and nmf was blantantly borrowed from the above post

#intialize the vectorizers
#this one is for lda
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
#below for nmf
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
```

```python
#clean data for nmf
tfidf_eap = tfidf_vectorizer.fit_transform(eap.text)
tfidf_feature_names_eap = tfidf_vectorizer.get_feature_names()

tfidf_hpl = tfidf_vectorizer.fit_transform(hpl.text)
tfidf_feature_names_hpl = tfidf_vectorizer.get_feature_names()

tfidf_mws = tfidf_vectorizer.fit_transform(mws.text)
tfidf_feature_names_mws = tfidf_vectorizer.get_feature_names()
```

```python
#clean data for lda
sent_clean = [[],[],[]]
feature_names = [[],[],[]]
t = 0
for series in [eap.text,hpl.text,mws.text]:
    tf = tf_vectorizer.fit_transform(series)
    tf_feature_names = tf_vectorizer.get_feature_names()
    
    sent_clean[t] = tf
    feature_names[t] = tf_feature_names
    t = t + 1
```

```python
#this is just for remebering where each author's senetecnes are
eap_clean = sent_clean[0]
hpl_clean = sent_clean[1]
mws_clean = sent_clean[2]
```

Some experimentation showed that 15 topics gave some plausible and helful results

```python
n_top = 15
```

```python
#Run lda for all the authors
lda_eap = LatentDirichletAllocation(n_topics=n_top, max_iter=5, learning_method='online',learning_offset=50.,random_state=0).fit(eap_clean)

lda_hpl = LatentDirichletAllocation(n_topics=n_top, max_iter=5, learning_method='online',learning_offset=50.,random_state=0).fit(hpl_clean)

lda_mws = LatentDirichletAllocation(n_topics=n_top, max_iter=5, learning_method='online',learning_offset=50.,random_state=0).fit(mws_clean)
```

```python
#Run nmf for all authors
nmf_eap = NMF(n_components=n_top, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf_eap)

nmf_hpl = NMF(n_components=n_top, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf_hpl)

nmf_mws = NMF(n_components=n_top, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf_mws)
```

```python
#this cell from https://medium.com/mlreview/topic-modeling-with-scikit-learn-e80d33668730
def get_topics(model, feature_names, no_top_words):
    topic_summ = []
    for topic_idx, topic in enumerate(model.components_):
        t_idx = '{0} '.format(topic_idx)
        s = t_idx + ' '.join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]])
        topic_summ.append(s)
    return(topic_summ)
```

I claim that nmf provides more human interpretable topics. Obviously, human interpretability is secondary for the author classification problem, but 
this is an oppurtunity to dicover which algorithm gets closer to human thought, as an intellectual exercise.

For LDA, topics 6 and 14 seem like reasonable images, with 6 a love of the physical country of england, and 14 as pwoerful feelings in te heart or soul.
Others are either indecipherable (topic 10), or complicated and difficult to qualify (topic 2)
NMF on the other hand, provides a whole slew of accurate and recognizable topics. These include the warm, positive emotions of topic 2, and 
then intense negativity of topic 14. Topics 5, 6, 12 and 13 also allow clear interpretation. Only three or so topics are complete nonsense.

```python
get_topics(lda_mws,feature_names[2],7)
```

```python
get_topics(nmf_mws,tfidf_feature_names_mws,7)
```

The difference in the algorithms is not so clear for eap; all his topics are either nonsense (topic 8 lda), or complex and difficult to decipher 
(topic 2 nmf).

```python
get_topics(lda_eap,feature_names[0],7)
```

```python
get_topics(nmf_eap,tfidf_feature_names_eap,7)
```

HPL's topics are, like EAP's, mostly nonsense, with an occasional topics which seem to conjure scenes, like topic 6 of the nmf.

```python
get_topics(lda_hpl,feature_names[1],7)
```

```python
get_topics(nmf_hpl,tfidf_feature_names_hpl,7)
```

## Visualizing the Topics

Unfortunately, the tool below only takes stochastic matrices (matrices where either the rows or columns sum to one, and can thus be interpreted as 
probablities), so we must restrict ourselves to vizulizing lda.

Although I prefer Non-negative Matrix Factorization for its tendency to give subjectively better results, LDA has a probability interpretation, which aids
description.

```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.disable_notebook()
# inspired by http://nbviewer.jupyter.org/github/bmabey/pyLDAvis/blob/master/notebooks/sklearn.ipynb
```

First, let's fit the models

```python
dtm_tf_eap = tf_vectorizer.fit_transform(eap.text)
lda_eap.fit(dtm_tf_eap)

dtm_tf_hpl = tf_vectorizer.fit_transform(hpl.text)
lda_hpl.fit(dtm_tf_hpl)

dtm_tf_mws = tf_vectorizer.fit_transform(mws.text)
lda_mws.fit(dtm_tf_mws)
```

Then, all we have to do is prepare the vizualization. We can display and save it later

```python
LDA_viz_eap = pyLDAvis.sklearn.prepare(lda_eap, dtm_tf_eap, tf_vectorizer)
LDA_viz_hpl = pyLDAvis.sklearn.prepare(lda_hpl, dtm_tf_hpl, tf_vectorizer)
LDA_viz_mws = pyLDAvis.sklearn.prepare(lda_mws, dtm_tf_mws, tf_vectorizer)
```

This tool not only provides a graph of the first two principal components, it also list the top 30 most salient terms overall, and for each topic. 
The equation used to generate "salience" is helpfully included in the bottom right hand corner. 

Hovering and or clicking on the topics provides not only that topics terms, but also the overall frequency overlaid with the frequency within that term.

In order to interpret the vizualization, a short description of PCA is neccesary. PCA finds and then orders the dimensions of a matrix so 
that the dimension which explains the data most is first, then the second most explanative dimension is second, etc.

Each topic is thus graphed on the first two dimensions. Topics that are near each other have a similar score on the two most important dimesnions. This is
evidence that the two topics have similar attributes, which in turn is evidence that the topics are similar.
